<h1>Section 4: Data Science</h1>

<h3>Professional Players vs Casual Players</h3>

<p>Being a professional player in League of Legends mean having to be the best of the best. Only the top <strong>0.01%</strong> of players in the rank ladder are taken into consideration by pro players. This makes a casual player like me ask: what makes a professional player stand out from casual players? Specifically, how different is the skill gap from professional and casual players?<br>

To answer this, I will be looking at two different roles within the game: <strong>the mid, and support</strong> roles. These positions are very different from each other, for example, a support player would (and should) have much less average kills than a mid laner. Likewise, a mid laner should have more average cs than a support. With these two roles, we will conduct a hypothesis test focusing on these two main roles and also create a multi-linear progression model.
</p>

<h2>Section 4.1: Mid laners</h2>

<p>First, let's conduct a hypothesis test on mid laners. CS is a stat that counts the amount of minions you have killed in a game. For a midlaner, being able to maintain high cs is very important. Having a high cs/min means having a steady income of gold throughout the game. Let's say that the average midlaner (which means gold / platinum players) gets about 6.5 cs per min. Therefore, the question we will be answering is 

> **Do pro players maintain significantly different amounts of cs/min compared to the average player?**

To answer this question, we will run a **one-sample-z-test**.

<h3>Hypothesis</h3>

My null hypothesis states that the average amount of sleep that pro players get is around 6.5 cs/min. The alternative hypothesis would be that pro players get more than 6.5 cs/min.

Cool! Now let's find some basic **summary statistics** before we conduct our z-test.
</p>

First, let's get our dataframe to only contain mid-laners:

```python
df_mid_laners = df[df["position"] == "mid"]
df_mid_laners
```

Now, using **df_mid_laners**, let's find the **mean cs/min** of all pro palyers, storing the result in the **variable** `cs_mean`:

```python
cs_mean = df_mid_laners["cspm"].mean()
cs_mean
```

Next, let's find the **standard deviation** of cs/min of pro mid laners, storing the result in the **variable** `cs_std`:

```python
cs_std = df_mid_laners["cspm"].std()
cs_std
```

Finally, find the **sample size** we're using for this z-test, storing the result in the **variable** `sample_size`. 

```python
sample_size = (df_mid_laners["cspm"]).count()
sample_size
```

Now that we've **calculated and stored** the necessary statistics, we can begin calculating the **p-value** and drawing conclusions about our hypotheses. For the purposes of this project, we will utilize a significance level of $0.05$. 

Let's calculate the **standard error** of the cs/min of pro mid laners, storing the result in the **numeric variable** `standard_error`. 

```python
import math
standard_error = cs_std/math.sqrt(sample_size)
standard_error
```

Next, **calculate our test statistic** given the `cs_mean` and `cs_error` found above, storing the result in the **numeric variable** `z_stat`. 

```python
z_stat = (cs_mean - 6.5)/standard_error
z_stat
```

Finally, we can calculate our p-value! Find it and store our p-value in the **variable** `p_value`.

```python
from scipy.stats import norm
p_value = (1 - norm.cdf(z_stat)) * 2
p_value
```

<h3>Analysis</h3>

<p>It looks like our p_value is so close to 0 that it comes out as 0. So, we **reject our null hypothesis** since our p-value is lower than our significance level (0.05). This means that the average pro midlaner has a higher cs/min than casual players.<br>

<h2>Section 4.2: Support players</h2>
That was pretty fun! Let's look at support players now! We need to create a new dataframe for pro support players.
</p>

```python
df_supports = df[df["position"] == "sup"]
df_supports
```

For support players, we will be conducting a **one-sided two-sample z-test**. 

Support players focus on assisting their team around the map rather than funneling gold for themselves. This means that support players will give cs to their other team members and prefer to deliver kills for their players, and racking up **assists**. However, another key part of League of Legends: while the map is usually empty, wards give a small amount of vision on the map. An important role for support players is placing these wards down around the map to give vision. This means that they want to have a high **vision score**.

We will be looking at these two factors to answer the question:

**Does buying more control wards mean having a higher *vision score* than casual support players?**

<h3>Hypothesis</h3>

My null hypothesis states that pro players who buy more control wards have a similar score. The alternative hypothesis would be that buying more control wards usually equates to having a vision score.

```python
df_ctrlwards_more = df_supports[df_supports["controlwardsbought"] >= 12]
df_ctrlwards_more
```

```python
df_ctrlwards_less = df_supports[df_supports["controlwardsbought"] < 12]
df_ctrlwards_less
```

```python
from statsmodels.stats.weightstats import ztest
ctrlward_results = ztest(df_ctrlwards_more["visionscore"], df_ctrlwards_less["visionscore"], alternative='larger')
ctrlward_results
```

<h3>Analysis</h3>

<p>It looks like our p_value is so close to 0 that it comes out as 0. So, we **reject our null hypothesis** since our p-value is lower than our significance level (0.05). This means that the average pro support has a higher vision score when buying more control wards.<br>

<h2>Section 4.3: Correlation between jungler assists / total cs and total gold</h2>
Let's find the correlation between multiple features of junglers such as assists, cs, total gold by using multi-linear regression with python<br>
First, let's create a dataframe for junglers:

```python
df_junglers = df[df["position"] == "jng"]
#df_junglers = df_junglers.dropna(axis=1, how="all")
df_junglers
```

Now let's only use the **assists**, **cs**, and **total gold** columns

```python
numerical_cols = ["assists", "total cs", "totalgold"]
df_num = df_junglers[numerical_cols]
df_num
```

Now let's use 10% of rows of df_junglers to train our model

```python
test_df = df_num.sample(frac=.1)
test_df
```

Now let's make our train-test split

```python
train_df = df_num[~df_num.index.isin(test_df.index)]
train_df
```

Now let's define our independent_variables and dependent_variable

```python
independent_variables = ["assists", "total cs"]
dependent_variable = "totalgold"
```

Now let's train the model

```python
# Importing the LinearRegression model from sklearn:
from sklearn.linear_model import LinearRegression

# Creating our model and storing in the variable `model`:
model = LinearRegression()

# Fitting the model to our **training** data. For almost all models in sklearn,
# the features (X) and labels (y) are passed as model.fit(X, y).
model.fit(train_df[independent_variables], train_df[dependent_variable])

# Generating predictions by passing our **testing** set features to model.predict()
predictions = model.predict( test_df[independent_variables] )
predictions
```
